```python
pH_level = survey_raw_df.pH.value_counts()
pH_level
```

```python
sns.countplot(y = survey_raw_df.pH)
plt.xticks(rotation=75);
```

```python
pH = survey_raw_df.groupby(['Nitrogen', 'potassium','Phosphorus']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
pH.plot.area()
plt.title('Soil Type')
plt.xlabel('pH level')
plt.ylabel('percentage ')
plt.legend(title='levels')
plt.show()
```
